```python
import pandas as pd
from kafka import KafkaProducer
from time import sleep
from json import dumps
import json
```

```python
producer = KafkaProducer(bootstrap_servers=['ip:9092'], #insert actual ip address here
                         value_serializer=lambda x:
                         dumps(x).encode('utf-8'))
```

```python
producer.send('sam_test', value = {'hello':'world'})
```
